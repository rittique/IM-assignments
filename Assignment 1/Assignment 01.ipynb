### Plotting the equations below and trying to find out their behaviours according to certain values

1. Logarithmic = θ(log n)
2. Linear = θ(n)
3. Quadratic = θ(n2)
4. Cubic = θ(n3)
5. Polynomial = θ(nk)

#### Input Values, n = 7, k = 4

```python
import matplotlib.pyplot as plt    
import numpy as np
```

```python
%matplotlib inline
```

```python
def compGraphs(n, k):
    
    x = np.linspace(1,n,50)
    
    log = np.log(x)
    linear = x
    quad = x ** 2
    cube = x ** 3
    poly = x ** k
    
    plt.plot(x, log, label = "Logarithmic")
    plt.plot(x, linear, label = "Linear")
    plt.plot(x, quad, label = "Quadratic")
    plt.plot(x, cube, label = "Cubic")
    plt.plot(x, poly, label = "Polynomial")
    
    fig = plt.gcf()
    fig.set_size_inches(18.5, 10.5)
    ax = fig.add_subplot(111)
    handles, labels = ax.get_legend_handles_labels()
    lgd = ax.legend(handles, labels, loc='upper left', bbox_to_anchor=(0.5,-0.1))
    #text = ax.text(-0.2,1.05, "Aribitrary text", transform=ax.transAxes)
    
    fig.savefig('graphs.png', dpi=300, bbox_extra_artists=(lgd,), bbox_inches='tight')
    
    
    plt.xlabel('x values', fontsize=16)
    plt.ylabel('y values',fontsize=16)
    plt.title('Plotted time complexity equations based upon Input', fontsize = 18)
    plt.legend(loc = "upper left")
    #plt.tight_layout()
    
    plt.show()
```

```python
compGraphs(7, 4)
```
